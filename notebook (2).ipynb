```python
hour_df1 = pd.read_csv(io.StringIO(hour_df['hour.csv'].decode('utf-8')))
print(hour_df1)
hour_df1.head()
```

membaca file hour.csv dan melihat 5 data teratas

```python
day_data = pd.read_csv(io.StringIO(day['day.csv'].decode('utf-8')))
print(day_data)
day_data.head()
```

membaca file day.csv dan melihat 5 data teratas

```python
gabungan_data = pd.merge(
    left = hour_df1,
    right = day_data,
    how = "left",
    left_on = "dteday",
    right_on = "dteday"
)
gabungan_data.head()
```

menggabungkan data hour.csv dengan day.csv menggunakan metode 'left' sehingga data pada hour.csv dapat digunakan dan dilengkapi dengan data dari day.csv

### Assessing Data

mengecek missing value. hasil pengecekan menunjukkan tidak adanya missing value

```python
gabungan_data['instant_x'] = gabungan_data['instant_x'].astype(str)
```

### Cleaning Data

```python
Q1 = (gabungan_data['cnt_y']).quantile(0.25)
Q3 = (gabungan_data['cnt_y']).quantile(0.75)
IQR = Q3 - Q1

maximum = Q3 + (1.5*IQR)
minimum = Q1 - (1.5*IQR)

kondisi_bawah = gabungan_data['cnt_y'] < minimum
kondisi_atas = gabungan_data['cnt_y'] > maximum

gabungan_data.drop(gabungan_data[kondisi_bawah].index, inplace = True)
gabungan_data.drop(gabungan_data[kondisi_atas].index, inplace = True)

gabungan_data.describe()
```

cleaning data dengan cara melihat data 'count' yang berupa outlier. tapi, dalam data ini ternyata tidak ada yang dihapus karena tidak ada data yang outlier

## Exploratory Data Analysis (EDA)

### Explore ...

```python
data2012 = gabungan_data[gabungan_data.yr_x == 1]
data2012.head()
```

membuat data frame untuk data 2012

```python
data1 = data2012.groupby(by = 'mnth_x').agg({
    'registered_y' : 'sum',
    'casual_y' : 'sum',
})
data1
```

```python
data2 = data2012.groupby(by = 'season_x').agg({
    'cnt_y' : 'sum',
})
data2
```

```python
data3 = data2012.groupby(by = 'weathersit_x').agg({
    'cnt_y' : 'sum',
})
data3
```

```python
data4 = data2012.groupby(by = 'mnth_x').agg({
    'cnt_y' : 'sum',
})
data4
```

```python
data2011 = gabungan_data[gabungan_data.yr_x == 0]
data2011.head()
```

```python
data5 = data2011.groupby(by = 'mnth_x').agg({
    'cnt_y' : 'sum',
})
data5
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
data1.plot(kind = 'bar', title = 'User per Month in 2012 (Million)', xlabel = 'Months in Year', label = ['registered user','casual user'])
```

```python
data2.plot(kind = 'pie', title = "Bike Rental per Season in 2012", labels = ['Spring', 'Summer', 'Fall', 'Winter'],
           autopct='%1.0f%%', legend = None, subplots = True)
plt.ylabel("")
```

```python
data3.plot(kind = 'bar', title = 'Weathersit', xlabel = 'Type of weather', ylabel = 'Bike rented', legend = None)
```

```python
plt.plot(data4, label = "2012")
plt.plot(data5, label = "2011")
plt.legend(loc = "upper left")
plt.title("Bike rented in 2011 and 2012")
plt.xlabel("Months")
plt.ylabel("Bike rented")
plt.show()
```
